```julia
using JuMP
using DelimitedFiles
using Cbc

#files = readdir("instancias/insta"; join=true)

#for f in files
model = Model(Cbc.Optimizer)
set_optimizer_attribute(model, "seconds", 60 * 30)
set_optimizer_attribute(model, "Heuristics", "off")
set_optimizer_attribute(model, "logLevel", 0)
set_optimizer_attribute(model, "threads", 8)
c = readdlm("instancias/insta/inst_1_1000.txt", Int)

n = size(c)[1]

print(c)

# Definindo variável de decisão -> Xij = {0,1}
@variable(model,x[i=1:n,j=1:n],Bin)

# Função objetivo
@objective(model, Max, sum(c[i, j] * x[i, j] for i in 1:n, j in 1:n))

# Restrições para garantir a saída de todo vértice
for i in 1:n
    @constraint(model,sum(x[i,j] for j in 1 : n) == 1)
end

# Restrições para garantir a chegada em cada vértice
for j in 1:n
    @constraint(model,sum(x[i,j] for i in 1:n) == 1)
end

# Trazendo uma solução otima
JuMP.optimize!(model)

# Checando se a solução é ótima ou não
if termination_status(model) == MOI.OPTIMAL
    optimal_solution = value.(x)
    optimal_objective = objective_value(model)
    println("Optimal solution -> ", optimal_objective)
elseif termination_status(model) == MOI.TIME_LIMIT && has_values(model)
    suboptimal_solution = value.(x)
    suboptimal_objective = objective_value(model)
    println("Suboptimal solution -> ", suboptimal_objective)
else
    error("The model was not solved correctly.")
end

#println("File name -> ",  f)
println("Agents -> ",  n)
println("Solve time -> ",  solve_time(model))
# println("Optimal solution -> ", solution(model))
println("\n-----------------------------\n")
    
#end
```

```julia
using JuMP
using Gurobi
using DelimitedFiles

const GRB_ENV = Gurobi.Env()

# Criando o modelo com o resolvedor Gurobi sem heurística e tempo limite de 30min
solver = () -> Gurobi.Optimizer(GRB_ENV)
```

```julia
# walkdir("instancias/insta")
files = readdir("instancias/insta"; join=true)

for f in files
    model = Model(solver)
    set_optimizer_attribute(model, "TimeLimit", 30)
    set_optimizer_attribute(model, "Heuristics", 0.0)
    set_optimizer_attribute(model, "OutputFlag", 0)
    
    c = readdlm(f, Int)

    n = size(c)[1]

    # Definindo variável de decisão -> Xij = {0,1}
    @variable(model,x[i=1:n,j=1:n],Bin)

    # Função objetivo
    @objective(model, Max, sum(c[i, j] * x[i, j] for i in 1:n, j in 1:n))

    # Restrições para garantir a saída de todo vértice
    for i in 1:n
        @constraint(model,sum(x[i,j] for j in 1 : n) == 1)
    end

    # Restrições para garantir a chegada em cada vértice
    for j in 1:n
        @constraint(model,sum(x[i,j] for i in 1:n) == 1)
    end
    
    # Trazendo uma solução otima
    optimize!(model)
    
    println("File name -> ",  f)
    println("Agents -> ",  n)
    println("Solve time -> ",  solve_time(model))
    println("\n-----------------------------\n")
    
end
```
